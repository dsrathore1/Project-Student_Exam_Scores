```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

#### **Load data from Processed data**

```python
load_data = pd.read_csv("../Data/processed/processed_v0.0.1.csv", na_values="-")
```

#### **Copy the data into new variable**

```python
df = load_data.copy()
```

#### **Name the index**

```python
df.index.name = "index"
df.head()
```

#### **Simple relation between Numeric Features using Pairplot**

```python
plt.figure(figsize=(10,8))
plt.suptitle("Relation Between Numeric Features")
sns.pairplot(df)
```

#### **Correlation between Numeric Data using Heatmap**

```python
plt.figure(figsize=(10,8))
plt.suptitle("Correlation between each numberic columns")
sns.heatmap(df.corr(), annot=True)
```

#### **Visualize the numeric columns using Histogram** 

```python
df.hist(figsize=(20,20), bins=50, color="purple")
plt.suptitle("Histograms of Numeric columns")
# plt.savefig("test1.png", dpi=400)
```

#### **Boxplot**

```python
plt.figure(figsize=(20,10))
sns.boxplot(
    data=df, orient="h", palette="Set2"
)  #! Orient is a parameter that specifies whether the plot is drawn vertically or horizontally (h/v).
```

#### **Scatter Plot for Hours Studied vs Exam Score**

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=df["hours_studied"], y=df["exam_score"], data=df)
plt.suptitle("Hours Studied vs Exam Score")
plt.show()
```

### **Line Plot Between Attendance vs Performance**

```python
plt.figure(figsize=(16,4))
sns.lineplot(x="attendance_percent", y="exam_score", data=df, color="purple", err_style="bars")
plt.suptitle("Attendance vs Performance")
# plt.savefig("Attendance vs Performance", dpi=500)
```
