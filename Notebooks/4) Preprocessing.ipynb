```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import StandardScaler, Normalizer
import joblib
```

### Load Dataset and Split the data

```python
load_data = pd.read_csv("../Data/processed/processed_v0.0.1.csv")

df = load_data.copy()
```

```python
X = df.drop(columns=["exam_score"])
Y = df["exam_score"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

### Scaling with StandardScaler

```python
scaler = StandardScaler()
X_train_scalered = scaler.fit_transform(X_train)
X_test_scalered = scaler.fit_transform(X_test)
```

```python
joblib.dump(scaler, "../Models/scaler_v0.1.pkl")
```

### Save scaler and processed sets

```python
pd.DataFrame(X_test_scalered, columns=X.columns).to_csv(
    "../Data/processed/Scalered/x_test_scalered.csv", index=False
)
pd.DataFrame(X_train_scalered, columns=X.columns).to_csv(
    "../Data/processed/Scalered/x_train_scalered.csv", index=False
)
Y_train.to_csv("../Data/processed/Scalered/y_train_scalered.csv", index=False)
Y_test.to_csv("../Data/processed/Scalered/y_test_scalered.csv", index=False)
```

### Scaling with Normalizer

```python
normalize = Normalizer()
X_train_normalized = normalize.fit_transform(X_train)
X_test_normalized = normalize.fit_transform(X_test)
```

```python
joblib.dump(normalize, "../Models/normalize_v0.1.pkl")
```

### Save Normalize and processed sets

```python
pd.DataFrame(X_test_normalized, columns=X.columns).to_csv(
    "../Data/processed/Normalized/x_test_normalized.csv", index=False
)
pd.DataFrame(X_train_normalized, columns=X.columns).to_csv(
    "../Data/processed/Normalized/x_train_normalized.csv", index=False
)
Y_train.to_csv("../Data/processed/Normalized/y_train_normalized.csv", index=False)
Y_test.to_csv("../Data/processed/Normalized/y_test_normalized.csv", index=False)
```
